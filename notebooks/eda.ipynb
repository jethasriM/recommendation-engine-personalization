```python
#STEP 1: Importing all the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings

warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] == (12,6)

print("All libraries successfully imported â˜‘ï¸")
```

```python
#STEP 2 Loading the dataset
print("...loading the rating csv ðŸ”ƒ")
ratings = pd.read_csv("../data/raw/ml-25m/ratings.csv")
print(f"loaded {len(ratings):,} ratings ")
print("...loading the movies.csv ðŸ”ƒ")
movies = pd.read_csv("../data/raw/ml-25m/movies.csv")
print(f"loaded {len(movies):,} movies")

print("Data loaded successfully!!ðŸŽ‰")
```

```python
#STEP 3 Basic Inspection
#Ratings Dataset
print("-"*60)
print(" "*20, "RATINGS DATASET", " "*20)
print("-"*60)
print("The shape of the ratings dataset:", ratings.shape)
print("\nThe columns:", list(ratings.columns))
print(f"\nThe first five values of ratings.csv:\n{ratings.head()}" )
print(f"\nThe last five values of ratings.csv:\n{ratings.tail()}" )
print(f"\nThe memory usage of ratings dataset:\n{ratings.memory_usage(deep=True).sum()/1024**2:.2f} MB")
#Movies Dataset
print(" ")
print("-"*60)
print(" "*20, "MOVIES DATASET", " "*20)
print("-"*60)
print("The shape of the ratings dataset:", movies.shape)
print("\nThe columns:", list(movies.columns))
print(f"\nThe first five values of ratings.csv:\n{movies.head()}", )
print(f"\nThe last five values of ratings.csv:\n{movies.tail()}" )
print(f"\nThe memory usage of ratings dataset:\n{movies.memory_usage(deep=True).sum()/1024**2:.2f} MB")
```

```python
#STEP 4: Basic Statistics
print("-"*60)
print(" "*20, "DATASET STATISTCS", " "*20)
print("-"*60)
n_ratings = len(ratings)
n_users = ratings['userId'].nunique()
n_movies = ratings['movieId'].nunique()

#DATASET SIZE
print(f"Number of Users:{n_users}")
print(f"Number of Movies:{n_movies}")
print(f"Number of ratings:{n_ratings}")

possible_ratings = n_users*n_movies
sparsity = (1-(n_ratings/possible_ratings))*100

print("-"*60)
print(" "*20, "MATRIX SPARSITY", " "*20)
print("-"*60)
print(f"Possible Ratings:{possible_ratings}")
print(f"Total number of actual ratings{n_ratings}")
print(f"The sparsity of the dataset:{sparsity:.4f}")
print(f"The density of the dataset:{(100-sparsity):.4f}")

print("-"*60)
print(" "*20, "RATING STATISTICS", " "*20)
print("-"*60)
print(f"{ratings['rating'].describe()}")
print(f" > Min:{ratings['rating'].min():.4f}")
print(f" > Max:{ratings['rating'].max():.4f}")
print(f" > Median:{ratings['rating'].median():.4f}")

#CHECK FOR MISSING VALUES
print("-"*60)
print(" "*20, "MISSING VALUES", " "*20)
print("-"*60)
print("\nThe number of missing values are:")
print(f"{ratings.isnull().sum()}")
```
